```python
import csv
import random
import math
```

```python
def load_dataset(filename):
    dataset = []

    with open(filename, "r") as file:
        reader = csv.reader(file)

        first_row = True

        for row in reader:
            if len(row) == 0:
                continue

            if first_row:
                first_row = False
                continue

            sample = []

            for value in row:
                try:
                    sample.append(float(value))
                except:
                    sample.append(value)

            dataset.append(sample)

    return dataset
```

```python
def split_dataset(dataset):
    train_set = []
    val_set = []
    test_set = []

    random.shuffle(dataset)

    for sample in dataset:
        r = random.random()

        if r <= 0.70:
            train_set.append(sample)
        elif r <= 0.85:
            val_set.append(sample)
        else:
            test_set.append(sample)

    return train_set, val_set, test_set
```

```python
def euclidean_distance(sample1, sample2):
    distance = 0

    for i in range(len(sample1) - 1):
        distance += (sample1[i] - sample2[i]) ** 2

    return math.sqrt(distance)
```

```python
def knn_classify(train_set, test_sample, k):
    distances = []

    for train_sample in train_set:
        dist = euclidean_distance(test_sample, train_sample)
        distances.append([dist, train_sample])

    distances.sort()

    neighbors = distances[:k]

    class_count = {}

    for neighbor in neighbors:
        label = neighbor[1][-1]

        if label not in class_count:
            class_count[label] = 0

        class_count[label] += 1

    predicted_class = max(class_count, key=class_count.get)

    return predicted_class
```

```python
def classification_accuracy(train_set, val_set, k):
    correct = 0

    for sample in val_set:
        predicted = knn_classify(train_set, sample, k)
        actual = sample[-1]

        if predicted == actual:
            correct += 1

    accuracy = (correct / len(val_set)) * 100

    return accuracy
```

```python
iris_data = load_dataset("iris.csv")

iris_train, iris_val, iris_test = split_dataset(iris_data)

k_values = [1, 3, 5, 10, 15]

best_k = None
best_accuracy = 0

print("K | Validation Accuracy")
print("----------------------")

for k in k_values:
    acc = classification_accuracy(iris_train, iris_val, k)
    print(k, "|", acc)

    if acc > best_accuracy:
        best_accuracy = acc
        best_k = k

print("Best K:", best_k)
```

```python
test_accuracy = classification_accuracy(iris_train, iris_test, best_k)

print("Test Accuracy:", test_accuracy)
```

```python
def knn_regression(train_set, test_sample, k):
    distances = []

    for train_sample in train_set:
        dist = euclidean_distance(test_sample, train_sample)
        distances.append([dist, train_sample])

    distances.sort()

    neighbors = distances[:k]

    total = 0

    for neighbor in neighbors:
        total += neighbor[1][-1]

    predicted_value = total / k

    return predicted_value
```

```python
def mean_squared_error(train_set, val_set, k):
    error = 0

    for sample in val_set:
        predicted = knn_regression(train_set, sample, k)
        actual = sample[-1]

        error += (actual - predicted) ** 2

    mse = error / len(val_set)

    return mse
```
